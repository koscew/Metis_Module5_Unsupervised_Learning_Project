## Web-scapring the list of games

```python
def game_list(top_n):
    '''
    enter numbers of top games to scrape top lists of games
    '''
    top_list = requests.get(f'https://boardgamegeek.com/browse/boardgame/page/{top_n//100}?sort=rank&sortdir=asc')
    top_list_soup = BeautifulSoup(top_list.content, "html5lib")
    rows = top_list_soup.find_all('tr', id='row_')
    columns = ['object_id','alias','board_game_rank','title','year','geek_rating','avg_rating','num_voters']
    df = pd.DataFrame(columns=columns)
    row_idx = 0
    for row in rows:
        cols = row.find_all('td')
        df.loc[row_idx] = [
            row.find_all('a')[1].get('href').split('/')[2],
            row.find_all('a')[1].get('href').split('/')[3],
            cols[0].text.strip(),
            cols[2].text.strip().split('\n\t\t\t\t\n\t\t\t\t\t\t\t')[0],
            cols[2].text.strip().split('\n\t\t\t\t\n\t\t\t\t\t\t\t')[1][1:5],
            cols[3].text.strip(),
            cols[4].text.strip(),
            cols[5].text.strip()]
        row_idx += 1  
    df.to_csv(f'data/top_{top_n}.csv')
    return f'list of top {top_n} games done'    
```

## Web-scraping the stats of games

```python
def single_game_stat(game_id):
    '''
    get the stats of a given game_id
    '''
    stats_re = requests.get(f'https://www.boardgamegeek.com/xmlapi/boardgame/{game_id}?stats=1')
    root = ET.fromstring(stats_re.text)
    stat_list = [game_id]
    
    stat_list.append(root.find('.//yearpublished').text)
    stat_list.append(root.find('.//minplayers').text)
    stat_list.append(root.find('.//maxplayers').text)
    stat_list.append(root.find('.//playingtime').text)
    stat_list.append(root.find('.//minplaytime').text)
    stat_list.append(root.find('.//maxplaytime').text)
    stat_list.append(root.find('.//age').text)
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[0].attrib['numvotes'])
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[1].attrib['numvotes'])
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[2].attrib['numvotes'])
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[3].attrib['numvotes'])
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[4].attrib['numvotes'])
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[5].attrib['numvotes'])
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[6].attrib['numvotes'])
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[7].attrib['numvotes'])
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[8].attrib['numvotes'])
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[9].attrib['numvotes'])
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[10].attrib['numvotes'])
    stat_list.append(root.findall('.//poll/[@name="suggested_playerage"]/results/result')[11].attrib['numvotes'])
    stat_list.append(root.find('.//statistics/ratings/usersrated').text)
    stat_list.append(root.find('.//statistics/ratings/owned').text)
    stat_list.append(root.find('.//statistics/ratings/numcomments').text)     
    return stat_list

def games_stat(game_list):
    '''
    get the stats of a given list of game_id
    '''
    columns = ['object_id', 'year', 'minplayers', 'maxplayers', 'playingtime', 'minplaytime', 'maxplaytime',
               'age', 'age_2', 'age_3', 'age_4', 'age_5', 'age_6', 'age_8', 'age_10', 'age_12', 'age_14',
               'age_16', 'age_18', 'age_21', 'usersrated', 'owned', 'numcomments']
    df = pd.DataFrame(columns=columns)
    row_idx = 0
    for g in game_list:
        df.loc[row_idx] = single_game_stat(g)
        time.sleep(5)
        row_idx += 1
    df.to_csv(f'data/{len(game_list)}_stat.csv')
    return f'stats of {len(game_list)} games done'  
```

## Web-scraping of comments

```python
def review_all(df_input):
    for g in df_input.object_id:
        numcomments = int(df_input.loc[df_input['object_id'] == g].numcomments)        
        for p in range(1, numcomments//100 + 1):
            review_re = requests.get(f'https://www.boardgamegeek.com/xmlapi/boardgame/{g}?comments=1&page={p}')
            root = ET.fromstring(review_re.text)
            df = pd.DataFrame(columns=['object_id', 'username', 'rating', 'comment'])
            row_idx = 0
            for comment in root.iter('comment'):
                df.loc[row_idx] = [
                    g,
                    comment.attrib['username'],
                    comment.attrib['rating'],
                    comment.text]
            df.to_csv(f'data/comment_{g}.csv')
            time.sleep(5)
    return f'comments of {g} games done'  

def review_page(game_list, page):
    for g in game_list:
        df = pd.DataFrame(columns=['object_id', 'username', 'rating', 'comment'])
        row_idx = 0
        for p in range(1, page+1):
            review_re = requests.get(f'https://www.boardgamegeek.com/xmlapi/boardgame/{g}?comments=1&page={p}')
            root = ET.fromstring(review_re.text)
            for comment in root.iter('comment'):
                df.loc[row_idx] = [
                    g,
                    comment.attrib['username'],
                    comment.attrib['rating'],
                    comment.text]
                row_idx += 1
            time.sleep(5) 
        df.to_csv(f'data/comment_{g}_{p}_pages.csv')
    return f'{page} pages of comments of {g} done'


def review_random(df_input, n, char):
    '''
    scraping random n over-char comments
    '''
    for g in df_input.object_id:
        df = pd.DataFrame(columns=['object_id', 'username', 'rating', 'comment'])
        row_idx = 0
        numcomments = int(df_input.loc[df_input['object_id'] == g].numcomments)
        pages = numcomments // 100 # + 1
        for p in np.random.randint(low=1, high=pages+1, size=pages):
            if row_idx < n:
                review_re = requests.get(f'https://www.boardgamegeek.com/xmlapi/boardgame/{g}?comments=1&page={p}')
                time.sleep(5) 
                root = ET.fromstring(review_re.text)
                for comment in root.iter('comment'):
                    if row_idx < n:
                        if len(comment.text) >= char:
                            df.loc[row_idx] = [
                                g,
                                comment.attrib['username'],
                                comment.attrib['rating'],
                                comment.text]
                            row_idx += 1
                        else:
                            continue
                    else:
                        break
            else:
                break                        
        df.to_csv(f'data/comments/comment_{g}_{n}.csv')
    return f'{n} comments of {g} done'
```
